```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# stocks = ['BAC', 'AXP', 'KO', 'VZ', 'MCO', 'USB', 'DVA', 'BK', 'CVX', 'VRSN', 'SHV']
# stocks = ['UAL', 'AAL', 'GE', 'SPY', 'XOM', 'JPM', 'F', 'CSCO', 'NKE', 'SHV']
# stocks = ['UAL', 'AAL', 'GE', 'SPY', 'XOM', 'JPM', 'F', 'CSCO', 'NKE', 'SHV']
stocks = ['UAL', 'AAL', 'GE', 'JPM', 'F', 'CSCO', 'NKE', 'XLY', 'XLP', 'XLE', 'XLF', 'XLV', 'XLI', 'XLB', 'XLK', 'XLU', 'SPY', 'SHV']
```

```python
info = pd.DataFrame(columns=['desc'])
for stock in stocks:
    info.loc[stock] = stock
```

```python
# Getting Price information
tickers = list(info.index.values)
df = yf.download(tickers, start="2009-04-30", end="2021-11-30")['Adj Close']
```

```python
df
```

```python
# Resampling to monthly data and changing dataframe to contain returns not share price.
prices = df.resample('M').last()
returns = prices.pct_change().iloc[1:,:]
returns
```

```python
# Getting Excess Returns (SHV is the monthly treasury yield)
excess = returns.subtract(returns['SHV'], axis=0)
excess = excess.drop(columns=['SHV'])
```

```python
with pd.ExcelWriter('industry_portfolio.xlsx') as writer:
    returns.to_excel(writer, sheet_name='total returns')
    excess.to_excel(writer, sheet_name='excess returns')
```
